# Exercício Velocidade

```python
import numpy as np
import matplotlib.pyplot as plt
import math as mt

%matplotlib inline

# PRIMEIRO TRECHO DO PERCURSO V = CTE=12
s = 5000   #distância
v = 12/3.6   #velocidade
t= s/v   # tempo total
#print(t)

N_t = int(t/60)  #numero de divisoes do tempo por 60

#Criar Lista

ls1 = np.linspace(0,s,N_t)
lt1 = np.linspace(0,t,N_t)
#print(ls1)

#SEGUNDO TRECHO ACELERADO
s2 = 200
v2= 15/3.6
a = (v2**2 -v**2)/s2
t2 = mt.sqrt(s2*2/a)
N_t2 = int(t2/0.1)


ls2 =np.linspace(s,s+(a/2)*t2**2,N_t2)
lt2 =np.linspace(t, t+t2, N_t2)


#print(ls2)
#print(lt2)

# TERCEIRO TRECHO V=CTE=15

s3= 2000
v3 = v2
t3 = s3/v3
N_t3 = int(t3/10)

ls3 = np.linspace(s+s2,s+s2+s3,N_t3)
lt3 = np.linspace(t+t2,t+t2+t3,N_t3)
#print(ls3)
#print(lt3)

#GRAFICO
#definindo os eixos
y = np.concatenate((ls1,ls2,ls3),axis=0)
x = np.concatenate((lt1,lt2,lt3),axis = 0)

#plot
fig, ax = plt.subplots()
ax.set_xlabel('t')
ax.set_ylabel('s(t)')
ax.plot(x , y)
```
