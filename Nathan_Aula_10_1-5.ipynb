# Exercício 3: Integral do Trapézio

```python

# Resolva Integral de Qq função usando o método do Trapézio/ Plote o Gráfico da Função
import numpy as np
import matplotlib.pyplot as plt
import math as mt
%matplotlib inline


A = float(input('A'))
B = float(input('B'))
N = 1200   #número de trapézios


#defina aqui a função a ser integrada
def func(x):
    return (x)  #f(x) escolha

#definindo o trapézio
def trapezoid(A,B,N): 
    h = (B-A)/(N-1) 
    soma = (func(A) + func(B))/2  
    for i in range(1,N-1):  #repetição para somar os trapézios
        soma += func(A+ i*h)
    return h*soma

print(trapezoid(A,B,N-1))

# Plot do grafico para o f(x)
x = np.linspace(A,B,N)
fig, ax = plt.subplots()
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.plot(x , func(x))
```

# Exercício 4: Integre Sin e Cos

```python

# Resolva Integral de Qq função usando o método do Trapézio/ Plote o Gráfico da Função
import numpy as np
import matplotlib.pyplot as plt
import math as mt
%matplotlib inline


A = float(input('A'))
B = 2*mt.pi
N = 15000   #número de trapézios


#defina aqui a função a ser integrada
def func(x):
    return (np.sin(x))  #f(x) escolha

#definindo o trapézio
def trapezoid(A,B,N): 
    h = (B-A)/(N-1) 
    soma = (func(A) + func(B))/2  
    for i in range(1,N-1):  #repetição para somar os trapézios
        soma += func(A+ i*h)
    return h*soma

print('Integral de Sin(x)=',trapezoid(A,B,N-1))

#defina aqui a função a ser integrada
def func2(x):
    return (np.cos(x))  #f(x) escolha

#definindo o trapézio
def trapezoid2(A,B,N): 
    h = (B-A)/(N-1) 
    soma2 = (func2(A) + func2(B))/2  
    for i in range(1,N-1):  #repetição para somar os trapézios
        soma2 += func2(A+ i*h)
    return h*soma2

print('Integral do Cos(x) =',trapezoid2(A,B,N-1))

# Plot do grafico para o f(x)
x = np.linspace(A,B,N)
fig, ax = plt.subplots(1,2,figsize=(9, 4.5), tight_layout=True )
ax[0].set_xlabel('x')
ax[0].set_ylabel('SIN(x)')
ax[0].plot(x , func(x), 'r-')

ax[1].set_xlabel('x')
ax[1].set_ylabel('COS(x)')
ax[1].plot(x , func2(x),'b-')
```
